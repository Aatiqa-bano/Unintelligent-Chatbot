```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("facebook/blenderbot-400M-distill")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/blenderbot-400M-distill")
```

```python
# Initialize conversation history as an empty list
conversation_history = []
```

```python
def get_response(input_text, conversation_history, max_history_length=6):
 # Limit conversation history to the last few messages
    truncated_history = conversation_history[-max_history_length:]
    history_string = "\n".join(truncated_history)

    # Tokenize input and history
    inputs = tokenizer(history_string + "\n" + input_text, return_tensors="pt", truncation=True)

    # Generate response from the model
    outputs = model.generate(**inputs, max_length=100)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

    # Update conversation history
    conversation_history.append(input_text)
    conversation_history.append(response)

    return response
```

```python
# Chat loop
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break

    try:
        bot_response = get_response(user_input, conversation_history)
        print(f"Chatbot: {bot_response}")
    except Exception as e:
        print(f"Error: {e}")
```
